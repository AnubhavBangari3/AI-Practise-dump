##Creating a random Tensor

```python
#Creating two random but same tensor
random_1=tf.random.Generator.from_seed(12)
print(random_1.normal(shape=[2,3]))

random_2=tf.random.get_global_generator()
print(random_2.normal(shape=[1,2]))
```

```python
#shuffle the order of elements in tensor
not_shuffle=tf.constant([[1,2],[4,5],[0,9]])
tf.random.shuffle(not_shuffle,seed=42)
```

```python
#random seed
tf.random.set_seed(12)
print(tf.random.uniform([1,1,2],seed=1))

```

```python
print(tf.random.set_seed(2))
print(tf.random.uniform([1,1,2],seed=1))
```

```python
#if want our tensor to be in same order
#use 1)global level tensor and 2) operational level tensor as mentioned above
```

##Following forms of aggregation
**Get the Minimum**
**Get the maximum**
**Get the mean of a tensor**
**Get the sum of a tensor**

```python
X=tf.constant([2,66,3,22])
Y=tf.constant([0,2,3,4])
tf.math.minimum(X,Y)
```

```python
tf.math.maximum(X,Y)
```

```python
x=tf.constant([[2,4],[0,8]])
tf.reduce_mean(x)
```

```python
tf.reduce_sum(x).numpy()
```

```python
tf.reduce_min(x).numpy()
```

```python
tf.reduce_max(x).numpy()
```

```python
import tensorflow_probability as tfp
tfp.stats.variance(x)
```

##Find the positional maxima and minimum

```python
tf.argmax(F).numpy()
```

```python
tf.argmin(F).numpy()
```

```python
F[tf.argmax(F)] == tf.reduce_max(F)
```
