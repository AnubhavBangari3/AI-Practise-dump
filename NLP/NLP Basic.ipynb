```python
import spacy
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
doc=nlp(u'Tesla is looking at buying US startup for $6 million')#NLP pipeline
```

```python
for token in doc:
  print(token.text,token.pos_,token.dep_)#text,part of speech,syntactic relation
```

**token.text:**This attribute represents the actual text (word or punctuation) of the token.


**token.pos_:** This attribute provides the part-of-speech (POS) tag of the token. POS tags are linguistic labels that categorize words based on their syntactic roles. Examples of POS tags include "NOUN" (noun), "VERB" (verb), "ADJ" (adjective), "ADV" (adverb), etc.


**token.dep_:** This attribute provides the dependency label of the token. In linguistic dependency parsing, words in a sentence are linked to each other based on grammatical relationships, and token.dep_ indicates the relationship of the token to its parent in the parse tree. Examples of dependency labels include "nsubj" (nominal subject), "dobj" (direct object), "prep" (prepositional modifier), etc.
