Non negative matrix Factorization.ipynb

```python
npr=pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/nlp udemy/npr.csv")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
t=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
```

```python
dtm=t.fit_transform(npr['Article'])
```

```python
dtm
```

```python
from sklearn.decomposition import NMF
```

```python
nmf_model=NMF(n_components=7,random_state=42)
```

```python
nmf_model.fit(dtm)
```

```python
t.get_feature_names_out()[2300]
```

```python
for i,t in enumerate(nmf_model.components_):
  print(f"The top  15 words for topic #{i}")
  print([ck.get_feature_names_out()[index] for index in t.argsort()[-15:]])
  print("\n")
```

```python
topic_results=nmf_model.transform(dtm)
```

```python
topic_results.argmax(axis=1)
```

```python
npr['Topic']=topic_results.argmax(axis=1)
```
