```python
doc6=nlp(u"Apple to built a startup with ANubhav Bangari for worth 60 million dollars")
```

```python
for i in doc6:
  print(i.text,end='|')
```

In spaCy, **ents** refers to the entities identified by the Named Entity Recognition (NER) component of the spaCy NLP pipeline. Named Entity Recognition is a common NLP task that involves identifying and classifying named entities in text into predefined categories such as persons, organizations, locations, dates, and more.

```python
for en in doc6.ents:
  print(en)
  print(en.label_)
  print("\n")
```

```python
doc7=nlp(u"Automatic car is not the future becuase Anubhav has said that.")
```

In spaCy, **"noun_chunks"** refers to the phrases or chunks of consecutive words that represent noun phrases in a sentence. Noun phrases are linguistic constructs that consist of a noun (or pronoun) and any words that modify it. They often function as the subject or object of a sentence.

```python
for i in doc7.noun_chunks:
  print(i)
```
