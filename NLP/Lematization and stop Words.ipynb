```python
import spacy
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
doc1=nlp(u"I am a runner runnin in the race becuase I love to run the race today")
```

```python
for token in doc1:
  print(token.text,'\t',token.pos_,'\t',token.lemma,'\t',token.lemma_)
```
