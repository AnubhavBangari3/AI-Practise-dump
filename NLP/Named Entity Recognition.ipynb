```python
import spacy
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
def show_ents(doc):
  if doc.ents:
    for ent in doc.ents:
      print(ent.text +'- '+ ent.label_ +'- '+str(spacy.explain(ent.label_)) )
  else:
    print("No entity found")
```

```python
doc=nlp(u"Hi how are you")
```

```python
show_ents(doc)
```

```python
doc=nlp(u"May I go to Kedarnath to get the blessings of lord shiva")
```

```python
show_ents(doc)
```

```python
doc=nlp("AB to build a INDIA factory for $6 million")
```

```python
show_ents(doc)
```

```python
from spacy.tokens import Span
```

```python
ORG=doc.vocab.strings[u"ORG"]
```

```python
ORG
```

```python
new_ent=Span(doc,0,1,label=ORG)
```

```python
doc.ents=list(doc.ents)+[new_ent]
#Adding our entity
```

```python
show_ents(doc)
```

```python
doc=nlp(u"Our company created an brand new vacuum cleaner."u"This new vacuum-cleaner is the best in show.")
```

```python
show_ents(doc)
```

```python
from spacy.matcher import PhraseMatcher
```

```python
matcher=PhraseMatcher(nlp.vocab)
```

```python
phrase_list=["vacuum cleaner","vacuum-cleaner"]
```

```python
phrase_patterns=[nlp(text) for text in phrase_list]
```

```python
matcher.add("newproduct",None,*phrase_patterns)
```

```python
found_matchers=matcher(doc)
```

```python
found_matchers
```

```python
from spacy.tokens import Span
```

```python
PROD=doc.vocab.strings[u'PRODUCT']
```

```python
new_ents=[Span(doc,match[1],match[2],label=PROD) for match in found_matchers]
```

```python
doc.ents=list(doc.ents)+new_ents
```

```python
show_ents(doc)
```
