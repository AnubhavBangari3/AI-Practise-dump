#Bag Of Words

```python
class CATEGORY:
  BOOKS="BOOKS"
  CLOTHING="CLOTHING"

train_X=["My fav book is Bhagwad Gita","I love this book","I love kurta","I love dress"]
train_Y=[CATEGORY.BOOKS,CATEGORY.BOOKS,CATEGORY.CLOTHING,CATEGORY.CLOTHING]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec=CountVectorizer()
v=vec.fit_transform(train_X)#Train X vectors
```

```python
print(vec.get_feature_names_out())
```

```python
print(v.toarray())
```

```python
from sklearn import svm
```

```python
c_svm=svm.SVC(kernel="linear")
c_svm.fit(v,train_Y)
```

```python
test_X=vec.transform(["My name is Anubhav Bangari"])
c_svm.predict(test_X)
```

```python
test_X=vec.transform(["People call me experience"])
c_svm.predict(test_X)
```

#Bigram, Ngrams

```python
vec_b=CountVectorizer(binary=True,ngram_range=(1,2))
v_b=vec_b.fit_transform(train_X)#Train X vectors
```

```python
print(vec_b.get_feature_names_out())
```

```python
print(v_b.toarray())
```

```python
c_svm=svm.SVC(kernel="linear")
c_svm.fit(v_b,train_Y)
```

```python
test_X=vec_b.transform(["My name is Anubhav Bangari"])
c_svm.predict(test_X)
```

#Word Vectors

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
docs=[nlp(text) for text in train_X]
train_X=[x.vector for x in docs]
print(docs[0].vector)
```

```python
c_svm_wv=svm.SVC(kernel="linear")
c_svm_wv.fit(train_X,train_Y)#33min in video
```

```python
test_x=["I love the book"]
test_docs=[nlp(text) for text in test_x]
test_x_v=[x.vector for x in test_docs]
c_svm_wv.predict(test_x_v)
```

```python
import re
```

```python
expr=re.compile(r"^ab[^\s]*cd$")
phrases=["abcd","xxxx","abxxxcd","ab cd"]
matches=[]
for x in phrases:
  if re.match(expr,x):
    matches.append(x)
print(matches)
```

#Stemming/Lemmatization

```python
import nltk
```

```python
nltk.download("wordnet")
nltk.download("stopwords")
```

```python
nltk.download("punkt")
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
phrase="reading the books"
words=word_tokenize(phrase)
stemmed_words=[]
for word in words:
  stemmed_words.append(stemmer.stem(word))

" ".join(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
stemmer=WordNetLemmatizer()
phrase="reading the books"
words=word_tokenize(phrase)
stemmed_words=[]
for word in words:
  stemmed_words.append(stemmer.lemmatize(word,pos='v'))

" ".join(stemmed_words)
```

#StopWords Removal

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
stop_words=stopwords.words('english')
print(stop_words)
```

```python
phrase="Anubhav Bangari is a millionaire. He is sanatani by birth and has pledged to save Hindustan"
words=word_tokenize(phrase)

ans=[]
for word in words:
  if word not in stop_words:
    ans.append(word)
" ".join(ans)
```
