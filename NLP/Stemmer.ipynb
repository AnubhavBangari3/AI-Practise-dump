```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

**PorterStemmer** refers to an implementation of the Porter stemming algorithm. The Porter stemming algorithm, developed by Martin Porter in 1980, is a widely used and popular algorithm for stemming English words.

**Stemming** is the process of reducing a word to its base or root form by removing suffixes and other affixes.

```python
words=['run','runner','ran','runs','easily','fairly','fairness']
```

```python
p_stemmer=PorterStemmer()
```

```python
for word in words:
  print(word+'------>'+p_stemmer.stem(word))
```

 Keep in mind that **stemming** might not always produce a meaningful word, as it relies on heuristic rules and may result in over-stemming or under-stemming in certain cases.

```python
from nltk.stem.snowball import SnowballStemmer
```

The **SnowballStemmer** in NLTK **is a stemming algorithm** that is based on the Snowball stemming algorithm, which was originally developed by Martin Porter (the same person who created the Porter stemming algorithm). The Snowball stemming algorithm is an improvement over the original Porter algorithm and is more versatile as it supports multiple languages.

```python
s_stemmer=SnowballStemmer(language='english')
```

```python
for word in words:
  print(word+'------>'+s_stemmer.stem(word))
```

```python
words2=["generous","generation","generously","generate"]
```

```python
for word in words2:
  print(word+'------>'+s_stemmer.stem(word))
```
