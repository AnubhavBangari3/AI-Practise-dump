### Select Specific Column

```python
df.select("name").show()
```

```python
df.select("name","salary").show()
```

```python
df.select(col("salary").alias("income")).show()
```

```python
df.selectExpr("salary*12 as annual_salary").show()
```

### Filter Rows

```python
df.filter(col("age")>10).show()
```

```python
df.filter((col("age")>10)&(col("city")=="Pune")).show()
```

```python
df.filter((col("age")>30)|(col("city")=="Pune")).show()
```

```python
df.filter(col("department").isin("AI","MLOps")).show()
```

### Add New Column

```python
df.withColumn("country",lit("India")).show()
```

```python
df.withColumn("bonus",col("salary")*0.2).show()
```

```python
df.withColumn("annualSalary",col("salary")*12).show()
```

```python
df.withColumn("level",when(col("experience")>5,"Senior").otherwise("Junior")).show()
```

### Rename Column

```python
df.withColumnRenamed("salary","monthly_salary").show()
```

### Delete Column

```python
df=df.withColumn("bonus",col("salary")*0.2)
```

```python
df.show()
```

```python
df.drop("bonus").show()
```

### Sort Data

```python
df.orderBy("salary").show()
```

```python
df.orderBy(col("salary").desc()).show()
```

```python
df.orderBy(col("salary"),col("department").desc()).show()
```

```python
df.orderBy(col("department"),col("salary").desc()).show()
```
