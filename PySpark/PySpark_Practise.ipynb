### Window Function

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *
```

```python
# Row Number (ascending salary within each department)
window_spec = Window.partitionBy("department").orderBy("salary")

df.withColumn("row_num", row_number().over(window_spec)).show()
```

```python
window_spec = Window.partitionBy("department").orderBy(desc("salary"))

df.withColumn("rank", rank().over(window_spec)).show()
```

```python
window_spec = Window.partitionBy("department").orderBy(desc("salary"))

df.withColumn("dense_rank", dense_rank().over(window_spec)).show()
```

```python
window_spec = Window.orderBy("salary")

df.withColumn("previous_salary", lag("salary", 1).over(window_spec)).show()
```

```python
window_spec = Window.orderBy("salary")

df.withColumn("next_salary", lead("salary", 1).over(window_spec)).show()
```
