```python
left=pd.DataFrame({
    'k':['k0','k1','k2','k3'],
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
})
right=pd.DataFrame({
    'k':['k0','k1','k2','k3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3'],
})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how='inner',on='k')
```

```python
pd.merge(left,right,how='left',on='k')
```

```python
pd.merge(left,right,how='right',on='k')
```

```python
left=pd.DataFrame({
    'k':['k0','k1','k2','k3'],
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3'],
},index=['k0','k1','k2','k3'])
right=pd.DataFrame({
    'k':['k0','k1','k2','k3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3'],
},index=['k0','k1','k2','k3'])
```

```python
left
```

```python
right
```
