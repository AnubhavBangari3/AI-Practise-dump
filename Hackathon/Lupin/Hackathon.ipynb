Handling missing Values

```python
train_data['liver_enzyme_lvl2'].fillna(train_data['liver_enzyme_lvl2'].median(), inplace=True)
train_data['smoking_habit'].fillna(train_data['smoking_habit'].mode()[0], inplace=True)
train_data['drinking_habit'].fillna(train_data['drinking_habit'].mode()[0], inplace=True)
```

```python
test_data['liver_enzyme_lvl2'].fillna(test_data['liver_enzyme_lvl2'].median(), inplace=True)
test_data['smoking_habit'].fillna(test_data['smoking_habit'].mode()[0], inplace=True)
test_data['drinking_habit'].fillna(test_data['drinking_habit'].mode()[0], inplace=True)
```

Handling categorical Values

```python
categorical_vals=['gender','can_hear_left_ear','can_hear_right_ear','smoking_habit','drinking_habit','residential_area']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
combined_data = pd.concat([train_data, test_data], keys=['train', 'test'])
combined_data = pd.get_dummies(combined_data, columns=categorical_vals)
train_data = combined_data.xs('train')
test_data = combined_data.xs('test')
```

Scaling

```python
train_data = train_data.drop(columns=['candidate_id'])
test_data_cand_id = test_data['candidate_id']
test_data = test_data.drop(columns=['candidate_id', 'triglyceride_lvl'])
```

```python
X = train_data.drop(columns=['triglyceride_lvl'])
y = train_data['triglyceride_lvl']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_after_scaling=scaler.fit_transform(X)
```

```python
test_data_scaled = scaler.transform(test_data)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_after_scaling, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
value_predictions = model.predict(X_val)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_val, value_predictions)
```

```python
print("mae:",mae)
```

```python
test_predictions = model.predict(test_data_scaled)
```

```python
print("test_predictions:",test_predictions)
```

```python
submission_data = pd.DataFrame({
    'candidate_id': test_data_cand_id,
    'triglyceride_lvl': test_predictions
})
```
