###Real world examples

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas=pd.read_csv('gas_prices.csv')
```

```python
gas
```

```python
plt.figure(figsize=(8,4))
#plt.plot(gas.Year,gas.USA,'b.-',label="USA")
#plt.plot(gas.Year,gas.Canada,'r.-',label="Canada")
#plt.plot(gas.Year,gas.Japan,'.-',label="Japan")
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],marker='.')
plt.xticks(gas.Year[::4])
plt.xlabel("Year")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
##FIFA
fifa=pd.read_csv("fifa_data.csv")
fifa
```

##Histograms

```python
bins=[40,50,60,70,80,90,100]
plt.hist(fifa.Overall,bins=bins)
plt.xticks(bins)
plt.xlabel("Skill level")
plt.ylabel("No of players")
plt.title("Players")
plt.legend()
plt.show()
```

##Pie chart


```python
left=fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
```

```python
labels=['left','right']
plt.pie([left,right],labels=labels,autopct='%.2f')
plt.title("Preferred foot")
plt.show()
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
fifa.Weight
```

```python
light=fifa.loc[fifa.Weight < 124].count()[0]
medium=fifa.loc[(fifa.Weight > 124) & (fifa.Weight < 150 )].count()[0]
heavy=fifa.loc[fifa.Weight > 150].count()[0]
```

```python
a=[light,medium,heavy]
labels=['light','medium','heavy']
explode=[.4,.3,.2]
plt.pie(a,labels=labels,autopct='%.2f',pctdistance=0.5,explode=explode)
```

```python
plt.title("Teams ")
barcelona=fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club == 'Real Madrid']['Overall']
team=[barcelona,madrid]
labels=['barcelona','madrid']
plt.boxplot(team,labels=labels)
plt.show()
```
