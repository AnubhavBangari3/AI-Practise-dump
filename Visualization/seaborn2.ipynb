```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)#for column vs col
```

```python
sns.PairGrid(iris)
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g.map_lower(sns.kdeplot)
```

```python
g.map_lower(sns.kdeplot)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
```

```python
tips=sns.load_dataset('tips')#category vs category
```

```python
tips.head()
```

```python
g=sns.FacetGrid(tips,row='sex',col='day')
```

```python
g=sns.FacetGrid(tips,row='sex',col='day')
g.map(plt.hist,'total_bill')
```

```python
g=sns.FacetGrid(tips,row='sex',col='day')
g.map(plt.scatter,'total_bill','tip')
```

```python
g=sns.JointGrid(x='total_bill',y="tip",data=tips)
```

```python
g.plot(sns.scatterplot, sns.histplot)
plt.show()
```
