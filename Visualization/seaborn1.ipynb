```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

##DISTRIBUTION PLOTS

```python
plt.figure(figsize=(12,9))
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],bins=50)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue="sex")
```

#CATEGORICAL DATA

```python
sns.barplot(x="sex",y="total_bill",data=tips)#x is categorical
```

```python
sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.mean)
```

```python
sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.std)
```

```python
sns.countplot(x="smoker",data=tips)
```

```python
sns.countplot(x="day",data=tips)
```

```python
sns.boxplot(x="sex",y="total_bill",data=tips)#dots are called outliers(possible extreme value)
#he line represents different quantile
```

```python
sns.boxplot(x="sex",y="total_bill",data=tips,hue="smoker")
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="smoker")#shows data distribution
 #these lines are kde
    #and whote dot is the median
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,hue="smoker",split=True)
```

```python
sns.violinplot(x="smoker",y="total_bill",data=tips,hue="day",split=True)#more thank two elements in  hue will give an error
```

```python
sns.stripplot(x="day",y="total_bill",data=tips)
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=False)#visualization on basis of density
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=False,hue="smoker")
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=False,hue="smoker",split=True)
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips)
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips,hue="sex")
```

```python

sns.swarmplot(x="day",y="total_bill",data=tips,hue="sex",split=True,color='b')
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",color='g')
```
